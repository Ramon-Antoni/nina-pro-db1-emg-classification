```python
# Loading TCN model weights from weights/tcn_model_weights.pth

import torch
from models.TCN import TCN

def load_trained_tcn(model_path):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Initialize your specific architecture
    model = TCN(num_inputs=10, num_channels=[64]*3, kernel_size=3)
    
    # Load weights
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    model.eval()
    print("TCN weights loaded successfully.")
    
    return model

# Setting relative file path
from pathlib import Path
project_root = Path.cwd()

# Usage
my_TCN_model = load_trained_tcn(project_root / "weights" / "tcn_model_weights.pth")

```

```python
# Testing the TCN
import torch
from models.TCN import test_loader

all_preds = []
all_labels = []

with torch.no_grad():  # Disable gradient calculation
    for batch_x, batch_y in test_loader:
        # Forward pass
        output = my_TCN_model(batch_x)
        
        # Get the index of the highest value (the predicted class)
        _, preds = torch.max(output, 1)
        
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(batch_y.cpu().numpy())

from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(all_labels, all_preds)
print(f"TCN Test Accuracy: {accuracy * 100:.2f}%")
```
